```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("Resume.csv")

print("Columns in dataset:", df.columns)
```

```python
possible_resume_cols = ["Resume_str", "resume", "Resume", "resume_text"]

resume_col = None
for col in possible_resume_cols:
    if col in df.columns:
        resume_col = col
        break

if resume_col is None:
    raise Exception("Resume column not found. Check dataset column names.")

print("Using resume column:", resume_col)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df['cleaned_resume'] = df[resume_col].apply(clean_text)
```

```python

job_description = """
Looking for a Python Machine Learning Engineer with experience
in NLP, Deep Learning, Scikit-learn, Pandas, and Data Analysis
"""

cleaned_jd = clean_text(job_description)
```

```python
documents = df['cleaned_resume'].tolist()
documents.append(cleaned_jd)

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

resume_vectors = tfidf_matrix[:-1]
job_vector = tfidf_matrix[-1]
```

```python
similarity_scores = cosine_similarity(resume_vectors, job_vector)

df['match_score'] = similarity_scores
```

```python

skills_db = [
    "python", "machine learning", "deep learning", "nlp",
    "tensorflow", "pandas", "numpy", "sql", "java",
    "react", "javascript", "data analysis"
]

def extract_skills(text):
    found_skills = []
    for skill in skills_db:
        if skill in text:
            found_skills.append(skill)
    return ", ".join(found_skills)

df['skills_found'] = df['cleaned_resume'].apply(extract_skills)
```

```python
ranked_df = df.sort_values(by="match_score", ascending=False)

print("\n========== TOP CANDIDATES ==========\n")

for i, row in ranked_df.head(5).iterrows():
    print("Category     :", row.get("Category", "N/A"))
    print("Match Score  :", round(row['match_score'], 3))
    print("Skills Found :", row['skills_found'])
    print("-" * 60)
```

```python
ranked_df.to_csv("screened_candidates.csv", index=False)
print("\nResults saved to screened_candidates.csv")
```

```python
import matplotlib.pyplot as plt

df['score'] = df['match_score']

viz_df = df.sort_values(by='score', ascending=False).head(10)

plt.figure()
plt.bar(viz_df.index.astype(str), viz_df['score'])
plt.xlabel("Candidate Index")
plt.ylabel("Match Score")
plt.title("Top Candidate Match Scores")
plt.show()
```
